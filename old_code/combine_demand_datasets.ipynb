# Australia Dataset (Canberra/NSW)

```python
import pandas as pd
import os
from tqdm import tqdm
from datetime import datetime, timedelta

dir_path = 'C:/Users/stucws/Documents/astar/sentiment-analysis/webscrape_headlines/external_data/aus_demand_data/'
files = os.listdir(dir_path)

for i, file_name in tqdm(enumerate(files)):
    if i == 0:
        df = pd.read_csv(dir_path + file_name)
    else:
        df = pd.concat([df, pd.read_csv(dir_path + file_name)], ignore_index=True)
        
df
```

```python
df["SETTLEMENTDATE"] = pd.to_datetime(df["SETTLEMENTDATE"])
df.set_index("SETTLEMENTDATE", inplace=True)
```

```python
df.index.nunique()
```

```python
test = df.iloc[1:].asfreq('1h')
test
```

```python
weather = pd.read_csv('../canberra_historical_hourly_weather.csv', index_col=0, parse_dates=True)
weather
```

```python
aqi = pd.read_csv("C:/Users/stucws/Documents/astar/sentiment-analysis/webscrape_headlines/external_data/aus_demand_data/Air_Quality_Monitoring_Data_20240904.csv")
aqi['DateTime'] = pd.to_datetime(aqi['DateTime'], format='mixed')
aqi.set_index('DateTime', inplace=True)
aqi.sort_index(inplace=True)
aqi.drop_duplicates(inplace=True)
aqi = aqi.pivot_table(index=aqi.index, columns='Name', values='AQI_Site')
aqi.columns = [col.lower() + '_aqi' for col in aqi.columns]
aqi['aggregate_aqi'] = aqi[aqi.columns].mean(axis=1)
aqi.dropna(inplace=True)
aqi
```

```python
aqi.to_csv('test.csv')
```

```python
import numpy as np

combined_df = pd.concat([test, weather, aqi], axis=1, join='inner')
combined_df.rename(columns={"TOTALDEMAND": "actual", "RRP": "price"}, inplace=True)
combined_df.drop_duplicates(inplace=True)
combined_df.drop(columns=['REGION', 'PERIODTYPE'], inplace=True)

print(len(combined_df))
combined_df = combined_df.asfreq('h')
print(len(combined_df)) # a lot of missing data in 2014 and 2024
combined_df = combined_df.loc['2015-01-01':'2023-12-31']
print(combined_df[combined_df.isna().any(axis=1)])
missing_dates = combined_df[combined_df.isna().any(axis=1)].index
object_columns = ['period', 'day_of_week', 'is_weekend', 'weatherDesc']
numeric_columns = [col for col in combined_df.columns if col not in object_columns]

for col in numeric_columns:
    combined_df[col] = combined_df[col].interpolate()
for col in object_columns:
    combined_df[col] = combined_df[col].ffill()
combined_df.loc['2021-03-11']
```

```python
import ast
combined_df['weatherDesc'] = combined_df['weatherDesc'].apply(lambda x: ast.literal_eval(x)[0]['value'])
combined_df
```

```python
combined_df.to_csv('../external_data/demand/demand_data_all_nsw.csv')
```

```python
# encode text features to numerical
numerical = combined_df
numerical['period'] = pd.Categorical(numerical['period']).codes
numerical['day_of_week'] = pd.Categorical(numerical['day_of_week']).codes
numerical['is_weekend'] = pd.Categorical(numerical['is_weekend']).codes
numerical['weatherDesc'] = pd.Categorical(numerical['weatherDesc']).codes
numerical
```
