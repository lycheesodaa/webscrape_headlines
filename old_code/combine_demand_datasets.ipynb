```python
import pandas as pd
import os
from tqdm import tqdm
from datetime import datetime, timedelta

dir_path = 'C:/Users/stucws/Documents/astar/data/demand_datasets/'
files = [file for file in os.listdir(dir_path) if file.endswith('.csv')]

dfs = []

for file in files:
    dfs.append(pd.read_csv(dir_path + file))
```

```python
for df in tqdm(dfs):
    df.rename(columns={df.columns[0]: "datetime"}, inplace=True)
    df["datetime"] = pd.to_datetime(df["datetime"], format='mixed')
    df.set_index("datetime", inplace=True)
    df.sort_index()
```

```python
dfs[2] = dfs[2][~dfs[2].index.duplicated(keep='first')]
```

```python
duplicated_indexes = dfs[2].index.duplicated(keep=False)
rows_with_repeated_indexes = dfs[2][duplicated_indexes]
rows_with_repeated_indexes
```

```python
for df in dfs:
    print(df.index.is_unique)
```

```python
combined_df = pd.concat(dfs, axis=1, join='inner')
combined_df
```

```python
combined_df.to_csv(dir_path + 'all/demand_data_all.csv')
```

```python
df = pd.read_csv(dir_path + 'all/demand_data_all.csv')
df
```

```python
df['USEP ($/MWh)'] = pd.concat([df['PRICE ($/MWh)'].iloc[:2831], df['USEP ($/MWh)'].iloc[2831:]])
df
```

```python
df.drop(columns=['DATE', 'PERIOD', 'TIME', 'LCP ($/MWh)', 'PRICE ($/MWh)', 'DEMAND (MW)', 'TCL(MW)', 'SOLAR(MW)', 'TCL (MW)'], inplace=True)
df
```

```python
df.columns
```

```python
rearranged = ['datetime', 'system_demand_actual', 'nem_demand_actual', 'nem_demand_forecast', 'USEP ($/MWh)', 'year', 'month', 'day', 'hour', 'period', 'day_of_week', 'is_weekend', 'humidity', 'tempC', 'heatIndexC', 'precipMM', 'windspeedKmph', 'winddirDegree', 'windGustKmph', 'weatherDesc', 'visibility', 'pressure', 'cloudcover', 'dewPointC', 'uvIndex', 'feelsLikeC', 'north', 'south', 'east', 'west', 'central']

len(rearranged) == len(df.columns)
```

```python
df = df[rearranged]
df.rename(columns={
    'system_demand_actual': 'actual',
    'nem_demand_actual': 'met',
    'nem_demand_forecast': 'forecast',
    'USEP ($/MWh)': 'price',
}, inplace=True)
df
```

```python
import ast
df['weatherDesc'] = df['weatherDesc'].apply(lambda x: ast.literal_eval(x)[0]['value'])
df
```

```python
df['aggregatePSI'] = df[['north', 'south', 'east', 'west', 'central']].mean(axis=1)
df
```

```python
df.to_csv(dir_path + 'all/demand_data_all_cleaned.csv', index=False)
```
